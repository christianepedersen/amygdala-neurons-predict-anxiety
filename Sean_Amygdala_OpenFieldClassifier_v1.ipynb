Sean "One Scruff" Piantadosi Inscopix amygdala imaging during open field data classification models

This data was collected through a miniature epifluorescence microscope that was mounted onto the head of a mouse. The animal was free to move throughout a 'Elevated Zero Maze' (EZM) behavioral arena while its amygdala neurons were imaged. The EZM is a large circular arena where the animal can either be in 1 of 2 types of regions: an open region where the animal does not have any cover and is exposed to any potential predators and a closed region where the animal has high walls on either side of it that provide it shelter. This type of behavioral arena is used to gauge how willing animals are to explore their environment and venture out into the regions in which they are exposed. Animals that are relatively unwilling to venture into the exposed regions are considered to display an "anxiety-like" phenotype.

Here I use the neural activity (fluorescence) of multiple amygdala neurons expressing GCaMP to predict whether an animal is in the exposed (open) or protected (closed) regions of the EZM arena.

### Using Recurrent Neural Networks (RNN) to consider a recent history of neural activity

We have seen through the use of Random Forest Classifiers and SVM Classifiers that the activity of amygdala neurons at a given moment can predict whether the animal is in a anxiety-inducing, exposed area or in a safe, covered area.

Now, we will use RNN to consider a 10 second history of neural activity rather than just the current time bin.

```python
# import packages

import tensorflow as tf
import keras
from keras import backend as K
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten , Dense, Activation,Dropout
from keras.layers.advanced_activations import LeakyReLU
```

```python
print(x_smote.shape)
print(y_smote.shape)
```

```python
### Plot time series data for some neurons

# specify columns to plot
groups = [12, 19, 26, 29, 34, 36]
i = 1
# plot each column
plt.figure()
for group in groups:
	plt.subplot(len(groups), 1, i)
	plt.plot(x_smote[:, group])
	plt.title(group, y=0.5, loc='right')
	i += 1
plt.show()
```

```python
# For each timestep, create a 2D matrix of historical values (for training data)
x_smote_num = x_smote#.values
y_smote_num = y_smote#.values
y_smote_num[y_smote_num==-1] = 0

timewindow = 5

steps = range(timewindow,x_smote_num.shape[0])

x_train_steps = np.zeros([len(steps),timewindow,x_smote_num.shape[1]])

y_train_steps = np.zeros(len(steps))

for timestep in steps:
    
    step_history = x_smote_num[range(timestep-timewindow,timestep),:]
    
    x_train_steps[timestep-timewindow,:,:] = step_history

    y_train_steps[timestep-timewindow] = y_smote_num[timestep]
    
```

```python
# For each timestep, create a 2D matrix of historical values (for testing data)
x_test_num = X_test#.values
y_test_num = y_test#.values
y_test_num[y_test_num==-1] = 0

timewindow = 5

steps = range(timewindow,x_test_num.shape[0])

x_test_steps = np.zeros([len(steps),timewindow,x_test_num.shape[1]])

y_test_steps = np.zeros(len(steps))

for timestep in steps:
    
    step_history = x_test_num[range(timestep-timewindow,timestep),:]
    
    x_test_steps[timestep-timewindow,:,:] = step_history

    y_test_steps[timestep-timewindow] = y_test_num[timestep]
    
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation

model = keras.models.Sequential()
model.add(keras.layers.LSTM(10,activation='relu',input_shape=x_train_steps.shape[-2:]))
model.add(Dense(1, input_shape=(10,)))
model.add(Activation('relu'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.metrics_names)

#print(model.summary())

batch_size = 5
epochs = 10

history = model.fit(x_train_steps, y_train_steps, batch_size=batch_size, epochs=epochs, verbose=1) #, validation_split=0.1)
score = model.evaluate(x_test_steps, y_test_steps, batch_size=batch_size, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# Plot confusion matrix for RNN classifier

from sklearn.metrics import confusion_matrix

rnn_pred = np.round(model.predict(x_test_steps))
rnn_pred[rnn_pred>1] = 1
#y_test_steps = y_test_steps.reshape(len(y_test_steps),1)

pd.DataFrame(
    confusion_matrix(y_test_steps, rnn_pred),
    columns=['Predicted Closed Arm', 'Predicted Open Arm'],
    index=['True Closed Arm', 'True Open Arm']
)
```
